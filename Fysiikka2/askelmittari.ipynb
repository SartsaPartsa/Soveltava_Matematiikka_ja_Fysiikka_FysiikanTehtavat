```python
# =====================================================
# üì± ASKELMITTARI - PhyPhox kiihtyvyysdatan analysointi
# =====================================================

# KIRJASTOT (ty√∂kalut joita tarvitsemme)
import numpy as np               # Matematiikkaa
import pandas as pd              # Excelin lukeminen
import matplotlib.pyplot as plt  # Kuvaajien tekeminen
from scipy.signal import butter, filtfilt, welch, find_peaks  # Signaalien k√§sittely
import os                        # Tiedostojen tarkastelu

# =====================================================
# ASETUKSET - Muuta n√§m√§ omiin tarpeisiisi!
# =====================================================
CSV_POLKU = "walk.csv"           # Oman CSV-tiedoston nimi
ASKELTAAJUUS_MIN = 0.5          # Hitain mahdollinen askeltaajuus (askelta/sekunti)
ASKELTAAJUUS_MAX = 4.0          # Nopein mahdollinen askeltaajuus (askelta/sekunti)

# Tarkistetaan l√∂ytyyk√∂ tiedosto
if not os.path.exists(CSV_POLKU):
    print(f"‚ö†Ô∏è VAROITUS: Tiedostoa '{CSV_POLKU}' ei l√∂ydy!")
    print(f"üìÇ Katsoin hakemistosta: {os.getcwd()}")
    print("üí° Tarkista tiedoston nimi tai siirr√§ tiedosto oikeaan paikkaan")

# =====================================================
# VAIHE 1: LUETAAN MITTAUSDATA TIEDOSTOSTA
# =====================================================
print("üîç Luetaan mittausdata...")

try:
    # Luetaan CSV-tiedosto (kuten Excel-taulukko)
    data = pd.read_csv(CSV_POLKU)
    print(f"‚úÖ Tiedosto luettu! Mittauksia: {len(data)} kpl")
    print(f"üìã Sarakkeet: {list(data.columns)}")
except:
    print("‚ùå Tiedoston luku ep√§onnistui! Tarkista tiedoston nimi.")
    raise

# =====================================================
# VAIHE 2: ETSIT√Ñ√ÑN AIKA JA KIIHTYVYYSARVOT
# =====================================================
print("üîç Etsit√§√§n aika ja kiihtyvyysarvot...")

# Etsit√§√§n aika-sarake (voi olla nimelt√§√§n "time", "t", "aika" jne.)
aika_sarakkeet = []
for sarake in data.columns:
    if "time" in sarake.lower() or "aika" in sarake.lower():
        aika_sarakkeet.append(sarake)

if not aika_sarakkeet:
    print("‚ùå Aikasaraketta ei l√∂ytynyt!")
    raise ValueError("Tiedostosta ei l√∂ytynyt aikaa")

aika_nimi = aika_sarakkeet[0]
aika = data[aika_nimi].values.astype(float)
print(f"‚è∞ K√§ytet√§√§n aikasaraketta: '{aika_nimi}'")

# Etsit√§√§n kiihtyvyyden X, Y, Z arvot
def etsi_sarake(hakusanat):
    """Etsii sarakkeen joka sis√§lt√§√§ jonkun hakusanoista"""
    for sarake in data.columns:
        sarake_pieni = sarake.lower()
        for sana in hakusanat:
            if sana in sarake_pieni:
                return sarake
    return None

x_nimi = etsi_sarake(["x", "acc x", "acceleration x"])
y_nimi = etsi_sarake(["y", "acc y", "acceleration y"])  
z_nimi = etsi_sarake(["z", "acc z", "acceleration z"])

if not all([x_nimi, y_nimi, z_nimi]):
    print("‚ùå Kiihtyvyyssarakkeiden l√∂yt√§minen ep√§onnistui!")
    print(f"L√∂ydetyt sarakkeet: {data.columns.tolist()}")
    raise ValueError("X, Y, Z sarakkeiden tunnistus ep√§onnistui")

print(f"üìä X-akseli: '{x_nimi}'")
print(f"üìä Y-akseli: '{y_nimi}'") 
print(f"üìä Z-akseli: '{z_nimi}'")

# Muutetaan kiihtyvyysarvot numeroiksi
x = data[x_nimi].values.astype(float)
y = data[y_nimi].values.astype(float)
z = data[z_nimi].values.astype(float)

# =====================================================
# VAIHE 3: LASKETAAN N√ÑYTTEENOTTOTAAJUUS
# =====================================================
# Taajuus = kuinka usein mittauksia otettiin sekunnissa
aikavalien_keskiarvo = np.median(np.diff(aika))  # Keskim√§√§r√§inen aika mittausten v√§lill√§
taajuus = 1.0 / aikavalien_keskiarvo  # Mittauksia per sekunti

print(f"‚öôÔ∏è Mittaustaajuus: {taajuus:.1f} Hz (mittausta/sekunti)")
print(f"‚è±Ô∏è Mittauksen kesto: {aika[-1]-aika[0]:.1f} sekuntia")

# =====================================================
# VAIHE 4: PIIRRET√Ñ√ÑN ALKUPER√ÑISET MITTAUKSET
# =====================================================
print("üìä Piirret√§√§n alkuper√§iset mittaukset...")

plt.figure(figsize=(12,6))
plt.plot(aika, x, label='X-suunta', alpha=0.8, linewidth=1)
plt.plot(aika, y, label='Y-suunta', alpha=0.8, linewidth=1)
plt.plot(aika, z, label='Z-suunta', alpha=0.8, linewidth=1)
plt.xlabel('Aika (sekuntia)')
plt.ylabel('Kiihtyvyys (m/s¬≤)')
plt.title('Alkuperaiset kiihtyvyysmittaukset')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# =====================================================
# VAIHE 5: ETSIT√Ñ√ÑN PARAS SIGNAALI ASKELEIDEN L√ñYT√ÑMISEEN
# =====================================================
print("üîç Etsit√§√§n parasta signaalia askeleiden tunnistukseen...")

# Lasketaan kokonaiskiihtyvyys (kaikkien suuntien yhteisvaikutus)
kokonais_kiihtyvyys = np.sqrt(x**2 + y**2 + z**2)

def etsi_paras_askel_signaali(signaalit, nimet, min_taajuus, max_taajuus):
    """
    Katsoo kaikki signaalit ja l√∂yt√§√§ sen jossa askelet n√§kyv√§t parhaiten.
    
    Miten toimii:
    1. Laskee jokaiselle signaalille "tehospektrin" (mitk√§ taajuudet ovat voimakkaita)
    2. Katsoo askeltaajuuden alueelta (0.5-4 Hz) mik√§ signaali on voimakkain
    3. Palauttaa parhaan signaalin nimen
    """
    paras_nimi = None
    paras_teho = -999999  # Iso negatiivinen luku
    paras_taajuus = None
    
    plt.figure(figsize=(12, 8))
    
    for i, (signaali, nimi) in enumerate(zip(signaalit, nimet)):
        # Laskee tehospektrin (Welch-menetelm√§)
        taajuudet, tehot = welch(signaali, fs=taajuus, nperseg=min(len(signaali), 2048))
        
        # Piirt√§√§ spektrin
        plt.subplot(2, 2, i+1)
        plt.loglog(taajuudet, tehot)
        plt.axvspan(min_taajuus, max_taajuus, alpha=0.3, color='red', 
                   label=f'Askelten alue {min_taajuus}-{max_taajuus} Hz')
        plt.xlabel('Taajuus (Hz)')
        plt.ylabel('Teho')
        plt.title(f'Tehospektri: {nimi}')
        plt.grid(True)
        plt.legend()
        
        # Katsoo vain askeltaajuuden aluetta
        alue_maski = (taajuudet >= min_taajuus) & (taajuudet <= max_taajuus)
        if alue_maski.sum() == 0:
            continue
            
        # Etsii suurimman tehon t√§lt√§ alueelta
        suurin_idx = np.argmax(tehot[alue_maski])
        alue_taajuudet = taajuudet[alue_maski]
        alue_tehot = tehot[alue_maski]
        
        # Jos t√§m√§ on paras t√§h√§n menness√§, tallentaa sen
        if alue_tehot[suurin_idx] > paras_teho:
            paras_teho = alue_tehot[suurin_idx]
            paras_nimi = nimi
            paras_taajuus = alue_taajuudet[suurin_idx]
    
    plt.suptitle('Etsitaan parasta signaalia askelille')
    plt.tight_layout()
    plt.show()
    
    return paras_nimi, paras_taajuus

# Tutkitaan kaikki signaalit
signaalit = [x, y, z, kokonais_kiihtyvyys]
signal_nimet = ['X-suunta', 'Y-suunta', 'Z-suunta', 'Kokonais']

paras_signaali_nimi, loytynyt_askeltaajuus = etsi_paras_askel_signaali(
    signaalit, signal_nimet, ASKELTAAJUUS_MIN, ASKELTAAJUUS_MAX)

print(f"üéØ Paras signaali askeleille: {paras_signaali_nimi}")
print(f"üö∂ L√∂ytynyt askeltaajuus: {loytynyt_askeltaajuus:.2f} Hz")
print(f"‚è±Ô∏è Se tarkoittaa {loytynyt_askeltaajuus*60:.0f} askelta minuutissa")

# =====================================================
# VAIHE 6: VALITAAN PARAS SIGNAALI JA SUODATETAAN SE
# =====================================================
print("üîß Valitaan paras signaali ja suodatetaan h√§ly pois...")

# Valitaan paras signaali
if paras_signaali_nimi == 'X-suunta':
    valittu_signaali = x
elif paras_signaali_nimi == 'Y-suunta':
    valittu_signaali = y
elif paras_signaali_nimi == 'Z-suunta':
    valittu_signaali = z
else:
    valittu_signaali = kokonais_kiihtyvyys

# M√§√§ritet√§√§n suodatusalue (pid√§mme vain askelten taajuudet)
ala_raja = max(0.4, loytynyt_askeltaajuus - 0.6)  # Ei liian matala
yla_raja = min(5.0, loytynyt_askeltaajuus + 0.8)  # Ei liian korkea
ala_raja = max(ala_raja, 0.4)  # Varmistetaan minimiraja

def suodata_signaali(signaali, taajuus, ala_hz, yla_hz):
    """
    Poistaa signaalista h√§ly√§ ja j√§tt√§√§ vain askelten taajuudet.
    
    Miten toimii:
    - Tekee "band-pass" suodattimen joka p√§√§st√§√§ l√§pi vain tietyn taajuusalueen
    - Poistaa hitaat liikkeet (alle 0.4 Hz)
    - Poistaa nopeat t√§rin√§t (yli 5 Hz) 
    - J√§tt√§√§ j√§ljelle vain askelten taajuudet
    """
    # Normalisoidaan taajuudet (tekninen vaatimus)
    nyquist = 0.5 * taajuus
    ala_norm = ala_hz / nyquist
    yla_norm = yla_hz / nyquist
    
    if ala_norm >= 1.0 or yla_norm >= 1.0:
        raise ValueError("Suodatusrajat liian korkeita!")
    
    # Luodaan Butterworth-suodatin ja k√§ytet√§√§n sit√§
    b, a = butter(4, [ala_norm, yla_norm], btype='bandpass')
    return filtfilt(b, a, signaali)

# Suodatetaan valittu signaali
suodatettu = suodata_signaali(valittu_signaali, taajuus, ala_raja, yla_raja)
print(f"üîß Suodatin: {ala_raja:.2f}-{yla_raja:.2f} Hz")

# N√§ytet√§√§n alkuper√§inen vs. suodatettu
plt.figure(figsize=(12,6))
plt.plot(aika, valittu_signaali, alpha=0.6, label=f'{paras_signaali_nimi} (alkuper√§inen)', linewidth=1)
plt.plot(aika, suodatettu, label='Suodatettu (puhdas)', linewidth=2, color='red')
plt.xlabel('Aika (sekuntia)')
plt.ylabel('Kiihtyvyys (m/s¬≤)')
plt.title('Signaalin puhdistus: alkuperainen vs. suodatettu')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# =====================================================
# VAIHE 7: LASKETAAN ASKELEET KAHDELLA TAVALLA
# =====================================================
print("üö∂ Lasketaan askeleet...")

# M√§√§ritet√§√§n askelten minimiet√§isyys (ei voi ottaa askeleita liian tihe√§√§n)
min_askel_aika = 0.28  # 0.28 sekuntia minimiet√§isyys askelten v√§lill√§
min_etaisyys_mittauksina = int(min_askel_aika * taajuus)

# TAPA 1: HUIPPUJEN LASKENTA
# Etsit√§√§n signaalin huiput (maksimikohdat)
kynnys = 0.2 * np.std(suodatettu)  # Kynnysarvo: 20% signaalin vaihtelusta

huiput, _ = find_peaks(suodatettu, 
                       distance=min_etaisyys_mittauksina,  # Minimiet√§isyys huippujen v√§lill√§
                       height=kynnys)                       # Minimikorkeus
askeita_huipuista = len(huiput)

# TAPA 2: NOLLAKOHDAT
# Lasketaan kuinka monta kertaa signaali menee miinuksesta plussaan
nollaylitykset = np.where((suodatettu[:-1] < 0) & (suodatettu[1:] >= 0))[0]

# Poistetaan liian l√§hell√§ olevat nollaylitykset
if len(nollaylitykset) > 0:
    etaisyydet_ok = np.r_[True, np.diff(nollaylitykset) > min_etaisyys_mittauksina]
    nollaylitykset = nollaylitykset[etaisyydet_ok]

askeita_nollakohdista = len(nollaylitykset)

print(f"üë£ TAPA 1 (huiput): {askeita_huipuista} askelta")
print(f"üë£ TAPA 2 (nollakohdat): {askeita_nollakohdista} askelta")

# N√§ytet√§√§n tunnistetut askeleet kuvassa
plt.figure(figsize=(12,6))
plt.plot(aika, suodatettu, label='Puhdas signaali', alpha=0.8, color='blue')
plt.plot(aika[huiput], suodatettu[huiput], 'ro', markersize=8, 
         label=f'Huiput ({askeita_huipuista} askelta)')

if len(nollaylitykset) > 0:
    plt.plot(aika[nollaylitykset], suodatettu[nollaylitykset], 'g^', markersize=6, 
             label=f'Nollakohdat ({askeita_nollakohdista} askelta)')

plt.axhline(y=kynnys, color='red', linestyle='--', alpha=0.7, 
           label=f'Kynnysarvo ({kynnys:.3f})')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3, label='Nollataso')

plt.xlabel('Aika (sekuntia)')
plt.ylabel('Kiihtyvyys (m/s¬≤)')
plt.title('Tunnistetut askeleet')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# =====================================================
# VAIHE 8: LASKETAAN ASKELTAAJUUS JA -NOPEUS
# =====================================================
mittauksen_kesto = aika[-1] - aika[0]
askeltaajuus = askeita_huipuista / mittauksen_kesto  # Hz (askelta/sekunti)
askelia_minuutissa = askeltaajuus * 60

print(f"‚è±Ô∏è Mittauksen kesto: {mittauksen_kesto:.1f} sekuntia")
print(f"üö∂ Askeltaajuus: {askeltaajuus:.2f} Hz ({askelia_minuutissa:.0f} askelta/min)")

# =====================================================
# VAIHE 9: TALLENNETAAN KAIKKI KUVAT
# =====================================================
print("üíæ Tallennetaan kuvat...")

# Kuva 1: Alkuper√§iset mittaukset
plt.figure(figsize=(12,6))
plt.plot(aika, x, label='X-suunta', alpha=0.8)
plt.plot(aika, y, label='Y-suunta', alpha=0.8)
plt.plot(aika, z, label='Z-suunta', alpha=0.8)
plt.xlabel('Aika (s)')
plt.ylabel('Kiihtyvyys (m/s¬≤)')
plt.title('Alkuper√§iset kiihtyvyysmittaukset')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('01_alkuperaiset_mittaukset.png', dpi=300, bbox_inches='tight')
plt.close()

# Kuva 2: Tehospektrit
plt.figure(figsize=(12, 8))
for i, (signaali, nimi) in enumerate(zip(signaalit, signal_nimet)):
    taajuudet, tehot = welch(signaali, fs=taajuus, nperseg=min(len(signaali), 2048))
    
    plt.subplot(2, 2, i+1)
    plt.loglog(taajuudet, tehot)
    plt.axvspan(ASKELTAAJUUS_MIN, ASKELTAAJUUS_MAX, alpha=0.3, color='red')
    plt.xlabel('Taajuus (Hz)')
    plt.ylabel('Teho')
    plt.title(f'Tehospektri: {nimi}')
    plt.grid(True)

plt.suptitle('Tehospektrit - etsitaan askeltaajuutta')
plt.tight_layout()
plt.savefig('02_tehospektrit.png', dpi=300, bbox_inches='tight')
plt.close()

# Kuva 3: Suodatus
plt.figure(figsize=(12,6))
plt.plot(aika, valittu_signaali, alpha=0.6, label=f'{paras_signaali_nimi} (alkuper√§inen)')
plt.plot(aika, suodatettu, label='Suodatettu', linewidth=2)
plt.xlabel('Aika (s)')
plt.ylabel('Kiihtyvyys (m/s¬≤)')
plt.title('Signaalin suodatus')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('03_suodatus.png', dpi=300, bbox_inches='tight')
plt.close()

# Kuva 4: Askeleiden tunnistus
plt.figure(figsize=(12,6))
plt.plot(aika, suodatettu, label='Suodatettu signaali', alpha=0.8)
plt.plot(aika[huiput], suodatettu[huiput], 'ro', markersize=8, 
         label=f'Huiput ({askeita_huipuista} askelta)')
if len(nollaylitykset) > 0:
    plt.plot(aika[nollaylitykset], suodatettu[nollaylitykset], 'g^', markersize=6, 
             label=f'Nollakohdat ({askeita_nollakohdista} askelta)')
plt.axhline(y=kynnys, color='red', linestyle='--', alpha=0.7, label='Kynnysarvo')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.xlabel('Aika (s)')
plt.ylabel('Kiihtyvyys (m/s¬≤)')
plt.title('Askeleiden tunnistus')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('04_askeleiden_tunnistus.png', dpi=300, bbox_inches='tight')
plt.close()

print("‚úÖ Kuvat tallennettu!")

# =====================================================
# üìã YHTEENVETO
# =====================================================
print("\n" + "="*60)
print("üìã YHTEENVETO - ASKELMITTARIN TULOKSET")
print("="*60)
print(f"üéØ Paras signaali askeleille: {paras_signaali_nimi}")
print(f"üîß K√§ytetty suodatin: {ala_raja:.2f}-{yla_raja:.2f} Hz")
print(f"üö∂ Askeita l√∂ytyi: {askeita_huipuista} (huiput) / {askeita_nollakohdista} (nollakohdat)")  
print(f"‚ö° Askeltaajuus: {askeltaajuus:.2f} Hz ({askelia_minuutissa:.0f} askelta/min)")
print(f"‚è±Ô∏è Mittauksen kesto: {mittauksen_kesto:.1f} sekuntia")
print("\nüìÇ Tallennetut kuvat:")
print("   ‚Ä¢ 01_alkuperaiset_mittaukset.png")
print("   ‚Ä¢ 02_tehospektrit.png")
print("   ‚Ä¢ 03_suodatus.png") 
print("   ‚Ä¢ 04_askeleiden_tunnistus.png")
print("="*60)
print("üí° VINKKI: Laske k√§sin askeleet videosta ja vertaa tuloksiin!")
print("   Esim: Jos k√§sin 50 askelta, kone 48 ‚Üí Virhe = (48-50)/50 = -4%")
```
